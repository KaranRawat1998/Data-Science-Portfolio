```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
sns.set_style('whitegrid')
sns.distplot(df['total_bill'],kde=False,color='r',bins=40)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')  #kind=kde,reg,hex
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.barplot(x='sex',y='total_bill',data=df,estimator=np.std) #by default estimator is meann
```

```python
sns.countplot(x='sex',hue='smoker',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
tc= df.corr()
```

```python
tc
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```
