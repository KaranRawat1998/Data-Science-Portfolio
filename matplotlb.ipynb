```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r-*')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('X&Y plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_title('BIG')
axes2=fig.add_axes([0.1,0.5,0.5,0.5])
axes2.plot(y,x,'r')
axes2.set_title('small')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

for ax in axes:
    ax.plot(x,y)
    
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig= plt.figure(figsize=(8,4))

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(8,8))

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig('new.jpg')
```

```python
fig= plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,label='x-y')
ax.plot(y,x,label='y-x')

ax.legend(loc=0) #for best position of legend
```
